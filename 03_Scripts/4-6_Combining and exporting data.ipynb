## 1. Importing libraries and dataframes

```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Importing the data frames

# creates a path with the source folder
path = r'C:\Users\Marc\Mariana_CareerFoundry'

# importing the csv file from original data
df_ords_prior = pd.read_csv(os.path.join(path, '02_Data', '2-1_Original', 'orders_products_prior.csv'), index_col = False)

# importing the orders.csv file from prepared data

df_ords = pd.read_csv(os.path.join(path, '02_Data', '2-2_Prepared', 'orders_checked.csv'), index_col = False)

df_prods = pd.read_csv(os.path.join(path, '02_Data', '2-2_Prepared', 'products_checked.csv'), index_col = False)
```

## 2. Merging df_ords and df_ords_prior using order_id

### 2.1. Checking dataframes

#### Checking df_ords_prior

```python
# checks shape
df_ords_prior.shape
```

#### Checking df_ords

```python
# checks shape
df_ords.shape
```

```python
df_ords.head()
```

```python
df_ords.columns
```

```python
df_ords = df_ords.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'])
df_ords.head()
```

### 2.2. Merging both df_ords and df_ords_prior on order_id

```python
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)

df_merged_large.head(20)
```

```python
df_merged_large['_merge'].value_counts()
```

### 2.3 Exporting new dataframe as 'orders_products_combined'

```python
# Export data to csv

df_merged_large.to_csv(os.path.join(path, '02_Data','2-2_Prepared', 'orders_products_combined.csv'))
```

```python
# Export data to pkl

df_merged_large.to_pickle(os.path.join(path, '02_Data','2-2_Prepared', 'orders_products_combined.pkl'))
```

## 3. Merging df_prods and df_ords

### 3.1. Checking Dataframes

#### Checking df_prods

#### Dataframe df_ords_prods

```python
# renaming the new created merged dataframe
df_ords_prods = df_merged_large
```

#### Getting rid of the column _merge 
There was a problem when applying the command "indicator" to a merge with a dataframe with an already existing column _merge

```python
# Drops the column _merge
df_ords_prods = df_ords_prods.drop(['_merge'], axis=1)
df_ords_prods.columns
```

```python
df_ords_prods.shape
```

```python
df_ords_prods.head()
```

### 3.2. Merges dataframes

```python
# Combining the dataframes

df_merged_all = df_ords_prods.merge(df_prods, on = 'product_id', indicator = True)
```

```python
df_merged_all.head(10)
```

```python
# Checks how it was done

df_merged_all['_merge'].value_counts()
```

### 3.3. Exports new dataframe as 'ords_prods_merge'

```python
# Exports data to csv

df_merged_all.to_csv(os.path.join(path, '02_Data','2-2_Prepared', 'ords_prods_merge.csv'))
```

```python
# Exports data to pkl

df_merged_all.to_pickle(os.path.join(path, '02_Data','2-2_Prepared', 'ords_prods_merge.pkl'))
```
