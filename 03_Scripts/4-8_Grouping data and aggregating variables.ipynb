```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

### Getting rid of the funky price values

```python
df_prods.head()
```

```python
# checks what are the maximum values that might be off

sorted_prices = sorted(df_prods['prices'], reverse=True)

print(sorted_prices[:20])
```

```python
# replacing the price values that are too big in the dataframe
df.replace(99999.0, 4.0, inplace=True) # replaces 99999.0 for 4 dollars (product is milk)
df.replace(14900.0, 4.0, inplace=True) # replaces 14900.0 for 4 dollars (product is milk)

# checking the dataframe
df['prices'].max()
```

## 2. Group the dataframe by department_id and order number

```python
# group by department id and aggregate by order number

df.groupby('department_id').agg({'order_number': ['mean']})
```

## 3. In comparison to the subset of 1 million rows there are some fluctuation around +/- 1 unit . Overall it does not seem to change any trend, but all the values span from 15 to 22, so 1 unit might make a difference on interpretation. If the computer power is available, it might be better to use the complete dataframe to have a more precise result.

```python
#group by department id and takes "max", "min" and "mean" values

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# create a max order column, the data is grouped by user_id
# the transform function generates the maximum order for each user

df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head(10)
```

## 4. Using loc function to create a 'loyalty flag'

```python
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df['loyalty_flag'].value_counts(dropna = False)
```

```python
df[['user_id', 'loyalty_flag', 'order_number']].head()
```

## 5. Basic statistics for the different loyalty categories

```python
#group by department id and takes "max", "min" and "mean" values

df.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### There is no significant difference between the mean values of the product prices purchased by the different categories

## 6. Creating a spender_flag

```python
# create a max order column, the data is grouped by user_id
# the transform function generates the maximum order for each user

df['mean_price'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df.loc[df['mean_price'] < 10, 'spender_flag'] = 'Low spender' # I had put 40 instead of 10 here, correted on 12.06.2024
```

```python
df.loc[df['mean_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
df['spender_flag'].value_counts(dropna = False)
```

## 7. Creating a frequent_flag

```python
# create a max order column, the data is grouped by user_id
# the transform function generates the maximum order for each user

df['frequency_customer'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df.loc[df['frequency_customer'] > 20, 'frequent_flag'] = 'Non-frequent customer'
```

```python
df.loc[(df['frequency_customer'] <= 20) & (df['frequency_customer'] > 10), 'frequent_flag'] = 'Regular customer'
```

```python
df.loc[df['frequency_customer'] <= 10, 'frequent_flag'] = 'Frequent customer'
```

```python
df['frequent_flag'].value_counts(dropna = False)
```
